```python
# List of columns to get unique values from
columns_of_interest = [
    'What are your primary goals or objectives when using a music playlist generation system based on free text input?',
    'How do you think an AI-based system creates a playlist?',
    'Based on what do you think a playlist should be created?',
    'Where do you search for new music?',
    'Which music listening platform you use?',
    'How often do you listen to music?',
    'What is your age range?',
    'What is your monthly income?',
    'How comfortable are you with using technology for music listening and discovery?',
    'How long have you been using the current music listening system?',
    'How comfortable are you using the current music streaming system?',
    'Are you satisfied with the music recommendations provided by these platforms?',
    'How often do you go out of your way to find new music?',
    'How satisfied are you with the current methods of discovering new music through existing platforms or services?',
    'Do you find it challenging to discover new music that matches your mood or preferences?',
    'How important is customization and personalization in your music listening experience?',
    'Would you be interested in a music playlist created based on what you describe in free text?',
    'How comfortable are you with describing your desired music playlist using free text input?',
    'Have you used any existing music playlist generation systems?',
    'Are there any privacy concerns or considerations you have regarding the use of your music listening data to improve playlist re'
]

```

```python
# Dictionary to store mappings
mappings = {}
```

```python
for column in columns_of_interest:
    unique_values = data[column].unique()
```

```python
# Create a mapping for each column and replace the values
for column in columns_of_interest:
    unique_values = data[column].unique()
    value_to_number = {value: number for number, value in enumerate(unique_values)}
    data[column] = data[column].replace(value_to_number)
    mappings[column] = value_to_number

# Display the DataFrame to verify changes
print(data.head())
```

```python

# Display the mappings for each column
for column, mapping in mappings.items():
    print(f"Column: {column}")
    print(mapping)
    print("\n")
```

```python
# Save the mappings dictionary to a JSON file
json_file_path = '/content/drive/MyDrive/mappings.json'
with open(json_file_path, 'w') as json_file:
    json.dump(mappings, json_file, indent=4)
```
